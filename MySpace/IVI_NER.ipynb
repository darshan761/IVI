```
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
is_noun = lambda pos: pos[:2] == 'NN'
lines = 'New risks emerge every day as markets get desrupted, political instability interrupts supply chains and new technology pushes boundaries across the risk landscape. '
tokenized = nltk.word_tokenize(akshay)
nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 

print(nouns)
```

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS


icons_predict = ['market.png', 'risk.png', 'political.png']

text = 'Risk is a much riskier proposition than it used to be. New risks emerge every day as markets get desrupted, political instability interrupts supply chains and new technology pushes boundaries across the risk landscape. Yet, while many organisations see risk as a negative, good risk management can actually help companies go faster'

from gensim.parsing.preprocessing import remove_stopwords
text = remove_stopwords(text)
print(text)

# str1 = ' '.join(str(e) for e in nouns)
text = "New risks emerge every day as markets get desrupted, political instability interrupts supply chains and new technology pushes boundaries across the risk landscape. "

def generate_wordcloud(text): # optionally add: stopwords=STOPWORDS and change the arg below
    wordcloud = WordCloud(
                          width=800, height=400,
                          relative_scaling = 1.0,
                          stopwords = {'to', 'of', 'as', 'get', 'and', 'the', 'new', 'across'} # set or space-separated string
                          ).generate(text)
    
    fig = plt.figure(1, figsize=(8, 4))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.savefig("WordCloud.png")
```

```
generate_wordcloud(text)
```
